```python
import gpxpy
```

```python
gpx_path = "/Users/eugeneleach/code/Eugle3/cycle_more/Lunch Ride.gpx"
with open(gpx_path, "r", encoding="utf-8") as f:
    gpx = gpxpy.parse(f)
```

```python
# Flatten all track points
coords = [
    (p.latitude, p.longitude)
    for track in gpx.tracks
    for segment in track.segments
    for p in segment.points
]
```

```python
print(f"Parsed {len(coords)} points")
print("First point:", coords[-1] if coords else None)
```

```python
len(coords)
```

```python
MAX_COORDS = 250
if len(coords) > MAX_COORDS:
    step = max(1, len(coords) // MAX_COORDS)
    coords = coords[::step] + [coords[-1]]
```

```python
len(coords)
```

```python
import openrouteservice
import os, json
```

```python
client = openrouteservice.Client(key=os.environ["AK"])
```

```python
route = client.directions(
        coordinates=coords,           # [[lon, lat], ...]
        profile="cycling-regular",    # or your preferred profile
        format="geojson",
        elevation=True,
        instructions=True,
        extra_info=["surface", "waytype", "waycategory", "steepness"],
    )
```
